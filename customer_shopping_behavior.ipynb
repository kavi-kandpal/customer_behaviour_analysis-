```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ',"_")
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
#create column age group lables
bins = [0, 25, 40, 60, 100]
labels = ['young', 'adult', 'young_adult', 'senior']
df['age_group']=pd.qcut(df['age'],4,labels=labels)
```

```python
df[["age","age_group","customer_id"]].head(20)
```

```python
df["frequency_of_purchases"].head(10)
```

```python
feq_mapping ={
     "Fortnightly":int(14),
     "Weekly":int(7),
    "Annually":int(365),
    "Quarterly":int(90)
}
df[('feqently_purchase')]=df['frequency_of_purchases'].map(feq_mapping).astype('Int64')
```
